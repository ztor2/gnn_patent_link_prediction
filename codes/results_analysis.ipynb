```python
import numpy as np
import pandas as pd
import pickle as pkl
import json
import scipy.sparse as sp
import networkx as nx
from networkx.readwrite import json_graph
from datetime import datetime, timedelta
import time
import random
from collections import OrderedDict
import warnings; warnings.filterwarnings('ignore')
```

```python
graphs = ['original', 'pred']
obj = []
for graph in graphs:
    with open('../data/'+graph +'.graph', 'r') as f:
        data = json.load(f)
    obj.append(data)
orig_g = json_graph.node_link_graph(obj[0])
pred_g = json_graph.node_link_graph(obj[1])
pd.DataFrame([[len(orig_g.nodes()),len(orig_g.edges())], [len(pred_g.nodes()),len(pred_g.edges())]],
             index=['original grapn', 'pred graph'], columns=['nodes', 'edges'])
```

### Degree centrality comparison

```python
orig_degcen = nx.degree_centrality(orig_g)
orig_degcen_sorted = {k: v for k, v in sorted(orig_degcen.items(), key=lambda item: -item[1])}
pd.DataFrame(orig_degcen_sorted.items(), columns=['nodeid', 'degcen']).iloc[:10]
```

```python
pred_degcen = nx.degree_centrality(pred_g)
pred_degcen_sorted = {k: v for k, v in sorted(pred_degcen.items(), key=lambda item: -item[1])}
pd.DataFrame(pred_degcen_sorted.items(), columns=['nodeid', 'degcen']).iloc[:10]
```
